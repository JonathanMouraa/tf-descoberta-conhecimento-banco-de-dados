```python
query_valor = {"valor": {"$gte": 90, "$lte": 200}}
atendimentos_valor = list(colecao_atendimentos.find(query_valor))
print(f"Atendimentos entre R$90 e R$200: {len(atendimentos_valor)}")
```

```python
df = pd.DataFrame(list(colecao_atendimentos.find()))
```

```python
nulos = df.isnull().sum()
print(f"\n Valores nulos encontrados:\n{nulos.to_string()}")
```

```python
faturamento_total = float(df['valor'].sum())
print(f"Faturamento Total: R$ {faturamento_total:.2f}\n")
```

```python
faturamento_por_cidade = {k: float(v) for k, v in df.groupby('cidade')['valor'].sum().items()}
print("Faturamento por Cidade:")
print(f"{faturamento_por_cidade}\n")
```

```python
faturamento_por_servico = {k: float(v) for k, v in df.groupby('servico')['valor'].sum().items()}
print("Faturamento por Serviço:")
print(f"{faturamento_por_servico}\n")
```

```python
ticket_medio_geral = float (df['valor'].mean())
print(f"Ticket Médio Geral: R$ {ticket_medio_geral:.2f}\n")
```

```python
ticket_medio_por_cidade = {k: float(v) for k, v in df.groupby('cidade')['valor'].mean().round(2).items()}
print("10. Ticket Médio por Cidade:")
print(f"{ticket_medio_por_cidade}\n")
```

```python
servico_mais_lucrativo = str(df.groupby('servico')['valor'].sum().idxmax())
valor_servico_lucrativo = float(df.groupby('servico')['valor'].sum().max())
print(f"Serviço mais lucrativo: {servico_mais_lucrativo} (R$ {valor_servico_lucrativo:.2f})\n")
```

```python
cidade_maior_faturamento = str(df.groupby('cidade')['valor'].sum().idxmax())
valor_cidade_faturamento = float(df.groupby('cidade')['valor'].sum().max())
print(f"Cidade com maior faturamento: {cidade_maior_faturamento} (R$ {valor_cidade_faturamento:.2f})\n")
```

```python
top_3_df = df.nlargest(3, 'valor').drop(columns=['_id'])
print("13. Top 3 Maiores Atendimentos:")
print(top_3_df.to_string(index=False))
top_3_atendimentos = top_3_df.to_dict(orient='records')
```

```python
relatorio = {
    "data_geracao": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "faturamento_total": faturamento_total,
    "faturamento_por_cidade": faturamento_por_cidade,
    "faturamento_por_servico": faturamento_por_servico,
    "ticket_medio_geral": round(ticket_medio_geral, 2),
    "ticket_medio_por_cidade": ticket_medio_por_cidade,
    "servico_mais_lucrativo": servico_mais_lucrativo,
    "cidade_maior_faturamento": cidade_maior_faturamento,
    "top_3_atendimentos": top_3_atendimentos
}
```

```python
colecao_relatorios.insert_one(relatorio)
print("\nRelatório salvo na coleção 'relatorios' do MongoDB!\n")
```
